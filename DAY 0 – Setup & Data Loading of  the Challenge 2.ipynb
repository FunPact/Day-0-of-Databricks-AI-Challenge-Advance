### Loading the November 2019 Data

```python
df_n = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/ecom/ecom_data/2019-Nov.csv")
)
```

```python
df_o = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/ecom/ecom_data/2019-Oct.csv")
)
```
